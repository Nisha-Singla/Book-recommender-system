## Popularity Based Recommender System

```python
# Merge ratings with books on 'ISBN' to include book details with each rating
ratings_with_name = ratings.merge(books,on="ISBN")
```

```python
# Count the number of ratings for each book
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()

# Rename the 'Book-Rating' column to 'num_ratings' for clarity
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)

# Display the resulting DataFrame
num_rating_df
```

```python
# Calculate the average rating for each book
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()

# Rename the 'Book-Rating' column to 'avg_ratings' for clarity
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

# Display the resulting DataFrame
avg_rating_df
```

```python
# Merge number of ratings and average ratings into a single DataFrame
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')

# Display the combined DataFrame
popular_df
```

```python
# Filter books with 250+ ratings and sort by highest average rating
popular_df = popular_df[popular_df['num_ratings'] >= 250].sort_values('avg_rating', ascending=False).head(50)
```

```python
# Adding book details like author and image, and keep only relevant columns
popular_df= popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

## Collaborative Filtering Based Recommender System

```python
# Finding users who have rated 200 or more books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >= 200
active_users = x[x].index
```

```python
# Keeping only ratings from active users (who rated 200+ books)
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
```

```python
# Finding books that got 50 or more ratings from active users
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
# Keeping only ratings for books that are rated 50+ times by active users
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0 , inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
from difflib import get_close_matches

def recommend(book_name):
    # Clean up input
    book_name = book_name.strip().lower()

    # Create a lowercase lookup map for better matching
    title_map = {title.lower(): title for title in pt.index}
    matches = get_close_matches(book_name, title_map.keys(), n=1, cutoff=0.6)

    if not matches:
        return "Book not found. Try another title or check spelling."

    matched_title = title_map[matches[0]]

    # Find the index of the matched book
    index = np.where(pt.index == matched_title)[0][0]
    similar_items = sorted(
        list(enumerate(similarity_score[index])),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    # Gather titles only
    recommended_titles = []
    for i in similar_items:
        temp_title = pt.index[i[0]]
        recommended_titles.append(temp_title)

    # Print each title on a new line
    for title in recommended_titles:
        print(f"{title}")
```
